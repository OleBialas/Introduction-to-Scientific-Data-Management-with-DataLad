## Introducing the DataLad Run Command

- clone penguins dataset
- Use datalad run with simple commands and scripts
- Check git log to see automatically generated messages

## Tracking Inputs and Outputs

- With the `--input` flag, datalad will automatically get the required inputs
- With the `--output` flag, datalad will automatically unlock the outputs so they can be overwritten

## Rerunning your Analysis

- Use `datalad rerun` and `datalad rerun since` to rerun one or multiple commands
- show that rerunning oly makes an entry in the history if there was a change
